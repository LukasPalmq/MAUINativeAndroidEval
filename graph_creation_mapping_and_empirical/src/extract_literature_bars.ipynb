```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import numpy as np
```

```python
authors = pd.read_excel('extracted_data.xlsx', usecols=['CPU', 'RAM', 'SIZE', 'FPS', 'GPU', 'NETWORK', 'BATTERY', 'RESP.TIMES', 'Task_Calculations', 
                                                        'Task_ListScrolling', 'Task_Sensors', 'Task_AppRelated', 'Task_Network', 'Task_Filesystem', 'Android', 'Android+iOS', 'iOS'], na_values="")
```

## Calculate the value count for each performance metric column:

```python

# Create value counts for each column
cpu = authors['CPU'].value_counts()
ram = authors['RAM'].value_counts()
size = authors['SIZE'].value_counts()
fps = authors['FPS'].value_counts()
gpu = authors['GPU'].value_counts()
network = authors['NETWORK'].value_counts()
battery = authors['BATTERY'].value_counts()
resp_times = authors['RESP.TIMES'].value_counts()

# Collect these values into a single dictionary but keep their keys separate
status = {'CPU': cpu, 'RAM': ram, 'SIZE': size, 'FPS': fps, 'GPU' : gpu, 'NETWORK' : network, 'BATTERY' : battery, 'RESPONSE TIME' : resp_times}

# Calculate the total counts for each category and sort them
counts_dict = {key: sum(status[key]) for key in status}
sorted_counts = sorted(counts_dict.items(), key=lambda item: item[1], reverse=True)

# Unzip the keys and values into two separate lists
keys, total_counts = zip(*sorted_counts)

# Plot a bar chart for each column showing the count of each value
plt.figure(figsize=(6, 3))  # Adjust the figure size to make the plot larger
ax = plt.subplot()
ax.yaxis.set_major_locator(ticker.MaxNLocator(integer=True))

bar = plt.bar(x=keys, height=total_counts, width=0.9)
ax.bar_label(bar, padding=3)
plt.xticks(rotation=70, ha='right')
plt.ylim(0, max(total_counts) * 1.1)  # Adjust the y-axis limits here
plt.savefig(fname='metric_counts.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

## Calculate the value count for each task column:

```python
# Create value counts for each task column
task_calculations = authors['Task_Calculations'].value_counts()
task_listscrolling = authors['Task_ListScrolling'].value_counts()
task_sensors = authors['Task_Sensors'].value_counts()
task_apprelated = authors['Task_AppRelated'].value_counts()
task_network = authors['Task_Network'].value_counts()
task_filesystem = authors['Task_Filesystem'].value_counts()

# Collect these values into a single dictionary but keep their keys separate
status = {'Task_Calculations': task_calculations, 'Task_ListScrolling': task_listscrolling, 'Task_Sensors': task_sensors, 'Task_AppRelated': task_apprelated, 'Task_Network' : task_network, 'Task_Filesystem' : task_filesystem}

# Calculate the total counts for each category and sort them
counts_dict = {key: sum(status[key]) for key in status}
sorted_counts = sorted(counts_dict.items(), key=lambda item: item[1], reverse=True)

# Unzip the keys and values into two separate lists
keys, total_counts = zip(*sorted_counts)

# Plot a bar chart for each column showing the count of each value
plt.figure(figsize=(6, 3))  # Adjust the figure size to make the plot larger
ax = plt.subplot()
ax.yaxis.set_major_locator(ticker.MaxNLocator(integer=True))

bar = plt.bar(x=keys, height=total_counts, width=0.9)
ax.bar_label(bar, padding=3)
plt.xticks(rotation=70, ha='right')
plt.ylim(0, max(total_counts) * 1.1)  # Adjust the y-axis limits here
plt.savefig(fname='task_counts.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

### "Bonus" value counts for operating systems:

```python
# Create value counts for each operating system column
android = authors['Android'].value_counts()
android_ios = authors['Android+iOS'].value_counts()
ios = authors['iOS'].value_counts()

# Collect these values into a single dictionary but keep their keys separate
status = {'Android': android, 'Android+iOS': android_ios, 'iOS': ios}

# Calculate the total counts for each category
total_counts = [sum(status[key]) for key in status])

# Plot a bar chart for each column showing the count of each value
plt.figure(figsize=(8, 6))  # Adjust the figure size to make the plot larger
ax = plt.subplot()
ax.yaxis.set_major_locator(ticker.MaxNLocator(integer=True))

bar = plt.bar(x=status.keys(), height=total_counts, width=0.9)
ax.bar_label(bar, padding=3)
plt.xticks(rotation=70, ha='right')
plt.savefig(fname='OS_target_counts.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import numpy as np
```

## Read the files and get the related data from each column

```python
MAUI = pd.read_excel('../data/MAUI.xlsx', usecols=['list_RAM_high', 'list_CPU_high', 'multi_RAM_high', 'multi_CPU_high', 'file_RAM_high', 'file_CPU_high', 'gps_RAM_high', 'gps_CPU_high', 'page_RAM_high', 'page_CPU_high', 'download_RAM_high', 'download_CPU_high'], na_values="")
native = pd.read_excel('../data/native.xlsx', usecols=['list_RAM_high', 'list_CPU_high', 'multi_RAM_high', 'multi_CPU_high', 'file_RAM_high', 'file_CPU_high', 'gps_RAM_high', 'gps_CPU_high', 'page_RAM_high', 'page_CPU_high', 'download_RAM_high', 'download_CPU_high'], na_values="")
```

```python
#scrolling list
list_CPU_high_MAUI = np.array(MAUI['list_CPU_high'].values)
list_CPU_high_MAUI = list_CPU_high_MAUI[~np.isnan(list_CPU_high_MAUI)]
list_CPU_high_native = np.array(native['list_CPU_high'].values)
list_CPU_high_native = list_CPU_high_native[~np.isnan(list_CPU_high_native)]
list_RAM_high_MAUI = np.array(MAUI['list_RAM_high'].values)
list_RAM_high_MAUI = list_RAM_high_MAUI[~np.isnan(list_RAM_high_MAUI)]
list_RAM_high_native = np.array(native['list_RAM_high'].values)
list_RAM_high_native = list_RAM_high_native[~np.isnan(list_RAM_high_native)]

#multicore
multi_CPU_high_MAUI = np.array(MAUI['multi_CPU_high'].values)
multi_CPU_high_MAUI = multi_CPU_high_MAUI[~np.isnan(multi_CPU_high_MAUI)]
multi_CPU_high_native = np.array(native['multi_CPU_high'].values)
multi_CPU_high_native = multi_CPU_high_native[~np.isnan(multi_CPU_high_native)]
multi_RAM_high_MAUI = np.array(MAUI['multi_RAM_high'].values)
multi_RAM_high_MAUI = multi_RAM_high_MAUI[~np.isnan(multi_RAM_high_MAUI)]
multi_RAM_high_native = np.array(native['multi_RAM_high'].values)
multi_RAM_high_native = multi_RAM_high_native[~np.isnan(multi_RAM_high_native)]

#file
file_CPU_high_MAUI = np.array(MAUI['file_CPU_high'].values)
file_CPU_high_MAUI = file_CPU_high_MAUI[~np.isnan(file_CPU_high_MAUI)]
file_CPU_high_native = np.array(native['file_CPU_high'].values)
file_CPU_high_native = file_CPU_high_native[~np.isnan(file_CPU_high_native)]
file_RAM_high_MAUI = np.array(MAUI['file_RAM_high'].values)
file_RAM_high_MAUI = file_RAM_high_MAUI[~np.isnan(file_RAM_high_MAUI)]
file_RAM_high_native = np.array(native['file_RAM_high'].values)
file_RAM_high_native = file_RAM_high_native[~np.isnan(file_RAM_high_native)]

#gps
gps_CPU_high_MAUI = np.array(MAUI['gps_CPU_high'].values)
gps_CPU_high_MAUI = gps_CPU_high_MAUI[~np.isnan(gps_CPU_high_MAUI)]
gps_CPU_high_native = np.array(native['gps_CPU_high'].values)
gps_CPU_high_native = gps_CPU_high_native[~np.isnan(gps_CPU_high_native)]
gps_RAM_high_MAUI = np.array(MAUI['gps_RAM_high'].values)
gps_RAM_high_MAUI = gps_RAM_high_MAUI[~np.isnan(gps_RAM_high_MAUI)]
gps_RAM_high_native = np.array(native['gps_RAM_high'].values)
gps_RAM_high_native = gps_RAM_high_native[~np.isnan(gps_RAM_high_native)]

#change page
page_CPU_high_MAUI = np.array(MAUI['page_CPU_high'].values)
page_CPU_high_MAUI = page_CPU_high_MAUI[~np.isnan(page_CPU_high_MAUI)]
page_CPU_high_native = np.array(native['page_CPU_high'].values)
page_CPU_high_native = page_CPU_high_native[~np.isnan(page_CPU_high_native)]
page_RAM_high_MAUI = np.array(MAUI['page_RAM_high'].values)
page_RAM_high_MAUI = page_RAM_high_MAUI[~np.isnan(page_RAM_high_MAUI)]
page_RAM_high_native = np.array(native['page_RAM_high'].values)
page_RAM_high_native = page_RAM_high_native[~np.isnan(page_RAM_high_native)]

#download file
download_CPU_high_MAUI = np.array(MAUI['download_CPU_high'].values)
download_CPU_high_MAUI = download_CPU_high_MAUI[~np.isnan(download_CPU_high_MAUI)]
download_CPU_high_native = np.array(native['download_CPU_high'].values)
download_CPU_high_native = download_CPU_high_native[~np.isnan(download_CPU_high_native)]
download_RAM_high_MAUI = np.array(MAUI['download_RAM_high'].values)
download_RAM_high_MAUI = download_RAM_high_MAUI[~np.isnan(download_RAM_high_MAUI)]
download_RAM_high_native = np.array(native['download_RAM_high'].values)
download_RAM_high_native = download_RAM_high_native[~np.isnan(download_RAM_high_native)]
```
